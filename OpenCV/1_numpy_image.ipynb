```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

%matplotlib inline
```

```python
pic = Image.open('D://python/udemy_cv/00-puppy.jpg')
```

```python
type(pic)
```

```python
pic_arr = np.asarray(pic)
```

```python
pic_arr.shape
```

```python
plt.imshow(pic_arr)
```

```python
# RGB
# Red channel values 0-255 (255: pure red)
pic_red = pic_arr[:,:,0]

# 0:black, 255:white (red)
plt.imshow(pic_red, cmap='gray') 
```

```python
# RGB
# Red channel values 0-255 (255: pure red)
pic_green = pic_arr[:,:,1]

# 0:black, 255:white (green)
plt.imshow(pic_green, cmap='gray') 
```

```python
pic_new = pic_arr.copy()
pic_new[:,:,0] = pic_new[:,:,1] = 0

# R:0, G:0.
plt.imshow(pic_new) 
```
