## BeautifulSoup를 통한 간단한 웹크롤링

```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get("http://www.cgv.co.kr/movies/?ft=0") # 홈페이지 접속
c = r.content  # content 받아옴
```

```python
c
```

```python
soup = BeautifulSoup(c, "html.parser")
```

```python
soup
```

```python
# ol ( li ( div/clss (strong)))
```

```python
ol = soup.find("ol")  # ol이라는 첫 번째(find 명령어: 1개) 태그를 찾아 변수에 저장
```

```python
ol
```

```python
li = ol.find_all("li") # ol에서 모든(find_all 명령어: 모두) li 태그들을 찾아 변수에 대입
```

```python
len(li) # 몇 갠지
```

```python
li[0] # 첫 번째 li 태그 출력
```

```python
li[1]
```

```python
li[2]
```

```python
for l in li:  # li에서 요소 하나씩 꺼내서 l에 대입
    div = l.find("div", {"class":"box-contents"})
    # div 태그 중 class 속성이 box-contents인 태그 검색 후 div에 대입
    strong = div.find("strong").text  # strong 태그의 텍스트를 변수 strong에 대입
    print(strong)
```

## 네이버 뉴스 크롤링

```python
# UI class=type06 headline ( li -> dl ( dt class=photo->photo , dt = link,title , dd=content,date,companay))
```

```python
r = requests.get("http://news.naver.com/main/list.nhn? mode=LSD&mid=sec&sid1=100")
c = r.content
```

```python
c
```

```python
soup = BeautifulSoup(c, "html.parser")
```

```python
soup
```

```python
all = soup.find("ul", {"class":"type06_headline"})
```

```python
all
```

```python
# 이미지 주소 가져오는 함수

def getImageSrc():
    dl = all.find_all("dl")      # dl 태그를 모두 찾는다
    for item2 in dl:
        try:                     # 에러가 없을 시(dl 태그 찾을 시)
            img = item2.find("dt", {"class":"photo"}).find("img")
                      # 찾은 dl 태그 중에서 dt 태그(class: photo)를 찾고, 다시 그 안에서 img 태그를 찾는다
            print(img['src'])
        except:                  # 에러 발생 시(dl 태그 존재하지 않을 시)
            print("No image")
```

```python
getImageSrc()
```

```python
# 주소와 기사제목 가져오는 함수
def getLinkAndTitle():
    dl = all.find_all("dl")
    for item2 in dl:
        link = item2.find("dt", {"class":""}).find("a")   
        print(link['href'])
        print(link.text.replace("\t", "").replace("\n", "")[1:len(link.text)+1])
        # replace를 통해 탭, 줄 띄움을 모두 공백으로 하고, 문장들 중 첫번 째 자리(스페이스)를 잘라낸다
```

```python
getLinkAndTitle()
```

```python
# 기사 내용을 가져오는 함수
def getContent():
    dl = all.find_all("dl")
    for item2 in dl:
        try:
            content = item2.find("dd")
            print(content.text.replace("\t","").replace("\n","").split("...")[0])
            print(content.find("span", {"class":"writing"}).text)
            print(content.find("span", {"class": "date"}).text)
        except:
            print("No Content")
```

```python
getContent()
```
