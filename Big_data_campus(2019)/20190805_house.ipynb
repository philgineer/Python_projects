```python
import numpy as np
import pandas as pd

train_row = pd.read_csv('train.csv')
test_row = pd.read_csv('test.csv')
```

```python
X_train_row = train_row.iloc[:, :-1]
X_train_row
```

```python
row = X_train_row.append(test_row, ignore_index=True)
```

```python
row
```

```python
row.info()
```

```python
data = row.copy()
```

```python
row.Alley.value_counts()
```

Alley: Type of alley access to property

       Grvl	Gravel
       Pave	Paved
       NA 	No alley access

```python
Alley_dummy = pd.get_dummies(row.Alley, prefix='Alley')
data = pd.concat([data, Alley_dummy], axis=1)
data
```

```python
data.drop('Alley', axis=1, inplace=True)
data.columns
```

```python
row.MiscFeature.value_counts()
```

MiscFeature: Miscellaneous feature not covered in other categories
		
       Elev	Elevator
       Gar2	2nd Garage (if not described in garage section)
       Othr	Other
       Shed	Shed (over 100 SF)
       TenC	Tennis Court
       NA	None

```python
MiscFeature_dummy = pd.get_dummies(row.MiscFeature, prefix='MiscFeature')
data = pd.concat([data, MiscFeature_dummy], axis=1)
data
```

```python
data.drop('MiscFeature', axis=1, inplace=True)
data.columns
```

```python
row.Fence.value_counts()
```

Fence: Fence quality
		
       GdPrv	Good Privacy
       MnPrv	Minimum Privacy
       GdWo	Good Wood
       MnWw	Minimum Wood/Wire
       NA	No Fence

```python
Fence_dummy = pd.get_dummies(row.Fence, prefix='Fence')
data = pd.concat([data, Fence_dummy], axis=1)
data
```

```python
data.drop('Fence', axis=1, inplace=True)
data.columns
```

```python
row.PoolQC.value_counts()
```

PoolQC: Pool quality
		
       Ex	Excellent
       Gd	Good
       TA	Average/Typical
       Fa	Fair
       NA	No Pool

```python
PoolQC_dummy = pd.get_dummies(row.PoolQC, prefix='PoolQC')
data = pd.concat([data, PoolQC_dummy], axis=1)
data
```

```python
data.drop('PoolQC', axis=1, inplace=True)
data.columns
```

```python
row.FireplaceQu.value_counts()
```

FireplaceQu: Fireplace quality

       Ex	Excellent - Exceptional Masonry Fireplace
       Gd	Good - Masonry Fireplace in main level
       TA	Average - Prefabricated Fireplace in main living area or Masonry Fireplace in basement
       Fa	Fair - Prefabricated Fireplace in basement
       Po	Poor - Ben Franklin Stove
       NA	No Fireplace

```python
FireplaceQu_dummy = pd.get_dummies(row.FireplaceQu, prefix='FireplaceQu')
data = pd.concat([data, FireplaceQu_dummy], axis=1)
data
```

```python
data.drop('FireplaceQu', axis=1, inplace=True)
data.columns
```

```python
data.info()
```

```python
data.drop('Id', axis=1, inplace=True)
data.info()
```
