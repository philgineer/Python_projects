## groupby

```python
# SQL groupby 명령어와 같음. split, apply, combine 과정을 거쳐 연산함
```

```python
import pandas as pd
from pandas import DataFrame

ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
         'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
         'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
         'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
         'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}

df = pd.DataFrame(ipl_data)
df
```

```python
df.groupby(['Team']).sum()
```

```python
df.groupby(['Team', 'Year']).sum()
```

```python
df.groupby(['Team', 'Year'])['Points'].sum()
```

```python
grouped = df.groupby('Team')

for name, group in grouped:
    print("Name:",name)
    print("Group:\n",group)
```

```python
grouped.get_group('Devils')
```

```python
# 추출된 group 정보에는 세 가지 유형의 apply 가능
# 1. Aggregation: 요약된 통계정보 추출
# 2. Transformation: 해당 정보 변환
# 3. Filtration: 특정 정보 제거해 보여주는 필터링 기능
```

```python
grouped.agg(sum)
```

```python
grouped.sum()
```

```python
grouped.mean()
```

```python
import numpy as np
grouped.agg(np.mean)
```

## feature scaling

```python
# Feature scaling (ppt p.450)
# 두 변수 중 하나의 값의 크기가 너무 큼. feature 간의 최대-최소 값의 차이를 맞춘다.
# 전략 1: Min-Max Normalization
# 전략 2: Standardization (Z-score Normalization) (실제 Min-Max 값 모를 때 활용 가능)
```

```python
df.groupby('Team').filter(lambda x: len(x) >= 3)  # filter 안에는 boolean 조건이 존재해야 함.
# len(x): grouped된 dataframe 개수
```
