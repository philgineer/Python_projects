# 내 정답

```python
def d(n):
    t=0
    for i in str(n):
        t+=int(i)
    return n+t
a=[]
for i in range(10000):
    a.append(d(i))
b=list(range(10000+1))
for i in sorted(set(b)-set(a)):
    print(i)
```
