## Fibonacci

```python
def fib(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib(n-1) + fib(n-2)
```

```python
for i in range(18):
    print(fib(i))
```

```python
n = int(input())
def fib(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib(n-1) + fib(n-2)
print(fib(n))
```

## TOP Chart codes

```python
def p(n):
    x=0
    y=1
    for i in range(n):
        x,y=y,x+y;
    return x
a=int(input())
print(p(a))
```

```python
def r(x):
    if x < 2: return x
    return r(x-1)+r(x-2)
print(r(int(input())))
```
